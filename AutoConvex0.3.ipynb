```python
site_avenue_conteudo = site_avenue.page_source
site_avenue_conteudo = BeautifulSoup(site_avenue_conteudo, 'html.parser')
```

```python
def str_para_float(texto):
    texto = texto.replace('U$ ','')
    texto = texto.replace('R$ ','')
    texto = texto.replace('.','')
    texto = texto.replace(',','')
    texto = texto.replace('≈ $','')
    texto = texto.replace(' ','')
    texto = float(texto)/100
    return texto
```

```python
ativos = site_avenue_conteudo.find_all('div', attrs={'class': 'ui card card-item'})

ativos_avenue = []

for ativo in ativos: 
    
    ativo_tag = ativo.find('p').text
    ativo_tag = ativo_tag.replace(' ','')
    
    ativo_preco = ativo.find('h3').text
    ativo_preco = str_para_float(ativo_preco)
    
    ativo_dados = ativo.find_all('strong')
    
    ativo_qtde = ativo_dados[0].text
    ativo_qtde = float(ativo_qtde)
    
    ativo_valor_aquisicao = ativo_dados[1].text
    ativo_valor_aquisicao = str_para_float(ativo_valor_aquisicao)
    
    ativo_valor_atual = ativo_dados[2].text
    ativo_valor_atual = str_para_float(ativo_valor_atual)
    
    ativos_avenue.append([ativo_tag, '', ativo_preco, ativo_qtde, ativo_valor_aquisicao, ativo_valor_atual])

    #print('Ativo: ', ativo_tag)
    #print('Preço: ', ativo_preco)
    #print('Quantidade: ', ativo_qtde)
    #print('Valor aquisição: ', ativo_valor_aquisicao)
    #print('Valor atual: ', ativo_valor_atual)
    #print(' ')
```

```python
percentual_meta = pd.read_excel(r'C:\Users\bsimoes\Documents\Person\CONVEX GLOBAL\Percentual_Meta.xlsx')
percentual_meta = percentual_meta.dropna()
percentual_meta = percentual_meta[['Ativo','%META']]
```

```python
relatorio_final = pd.merge(relatorio, percentual_meta, how= 'left', on= 'Ativo')
```

```python
#display(percentual_meta)
display(relatorio_final)
```

```python
patrimonio_total = patrimonio_avenue_usd + patrimonio_binance_usd
display(patrimonio_total)
```

```python
relatorio_final['USD Meta'] = relatorio_final['%META']*patrimonio_total
```

```python
relatorio_final['Ação'] = [None]*len(relatorio_final)
```

```python
for i in range(len(relatorio_final)):
    if relatorio_final['USD Meta'][i] > relatorio_final['Valor atual'][i]:
        relatorio_final['Ação'][i] = 'Comprar'
    elif relatorio_final['USD Meta'][i] < relatorio_final['Valor atual'][i]:
        relatorio_final['Ação'][i] = 'Vender'
    else:
        relatorio_final['Ação'][i] = 'Manter'
```

```python
relatorio_final['USD Var'] = relatorio_final['Valor atual'] - relatorio_final['USD Meta']
relatorio_final['% var'] = relatorio_final['USD Var']/relatorio_final['USD Meta']
```
